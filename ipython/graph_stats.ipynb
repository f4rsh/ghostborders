```python
import numpy as np
data = np.genfromtxt('/Users/telmo/projects/ghostborders/belgium_dist_time.csv', names=['dist', 'time'], skip_header=1, delimiter=',')
```

```python
%matplotlib inline

import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

# the histogram of the data
n, bins, patches = plt.hist(data['dist'], 50)

plt.xlabel('Distance')
plt.ylabel('Count')
plt.title('Graph link distance histogram')
#plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

```python
# the histogram of the data
n, bins, patches = plt.hist(data['time'], 50)

plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Graph link time histogram')
plt.grid(True)

plt.show()
```

```python
#plt.scatter(data['dist'], data['time'], alpha=0.5, s=1.)
plt.hexbin(data['dist'], data['time'], xscale='log', yscale='log')
plt.show()
```
