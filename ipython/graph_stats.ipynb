```python
import numpy as np
data = np.genfromtxt('/Users/telmo/projects/ghostborders/belgium-dists.csv', names=['dist',])
```

```python
%matplotlib inline

import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(data['dist'], 50)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Distance')
plt.ylabel('Count')
plt.title('Graph link distance histogram')
#plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```
