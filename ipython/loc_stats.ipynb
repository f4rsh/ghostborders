```python
import numpy as np
data = np.genfromtxt('/Users/telmo/projects/ghostborders/belgium2-locs-metrics.csv', names=['loc','lat','lng','photos','users','dist','time','degree','neighbors','self','entropy','dist_var','angle_var','angle_entropy'], skip_header=1, delimiter=',')
data = data[np.logical_not(data['angle_entropy'] <= 0)]
```

```python
%matplotlib inline

import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

# the histogram of the data
a = np.array(data['degree'])
n, bins, patches = plt.hist(a[a != 0], 50)

plt.xlabel('Distance')
plt.ylabel('Count')
plt.title('Graph link distance histogram')
#plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

```python
# the histogram of the data
n, bins, patches = plt.hist(data['angle_entropy'], 50)

plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Graph link time histogram')
plt.grid(True)

plt.show()
```

```python
x = np.array(data['dist'])
y = np.array(data['angle_entropy'])
plt.hexbin(x, y, xscale='log', yscale='log', bins='log')
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x = 'degree'
y = 'angle_entropy'
z = 'dist'

X = data[x]
Y = data[y]
Z = data[z]

X = np.log10(data[x])
Y = np.log10(data[y])
Z = np.log10(data[z])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z)

ax.set_xlabel(x)
ax.set_ylabel(y)
ax.set_zlabel(z)

plt.show()
```
